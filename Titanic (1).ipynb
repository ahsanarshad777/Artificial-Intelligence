```python
import pandas as pd
import numpy as np
dataFrame = pd.read_csv (r'tested.csv')
```

```python
dataFrame
```

#### **Before** imputing Total **null values (missing Values)**

```python
na = dataFrame.isna( ).sum()
na
```

#### **Removing Cabin Column in Dataset.**

```python
dataFrame = dataFrame.drop(columns= 'Cabin', axis=1)
dataFrame
```

### **Taking Care of missing Data**

#### Imputing **Age** And **Fare** with **Mean Values**

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(dataFrame[['Age','Fare']])
dataFrame[['Age','Fare']] = imputer.transform(dataFrame[['Age','Fare']])
```

```python
dataFrame[['Age','Fare']]
```

#### **After** imputing Total **null values (missing Values)**

```python
na = dataFrame.isna( ).sum()
na
```

```python
dataFrame
```

#### **Converting Age column float to integer**

```python
dataFrame['Age'] = dataFrame['Age'].astype(int)
dataFrame['Age'] 
```

### **How many passengers survived and how many were killed (in number and %)**

```python
Total_survived= dataFrame[['Survived', 'PassengerId']]   
ds= Total_survived.groupby(['Survived'])
```

##### Survived & Killed In **Numbers**

```python
total_survived_in_Number = ds.count()
total_survived_in_Number
```

#### Killed In **Percent**

```python
total_killed_in_Percent = (total_survived_in_Number['PassengerId'][0]/total_survived_in_Number['PassengerId'].sum())*100
print('Total',total_killed_in_Percent.round(), '% were killed in Titanic.')
```

#### Survived In **Percent**

```python
total_survived_in_Percent = (total_survived_in_Number['PassengerId'][1]/total_survived_in_Number['PassengerId'].sum())*100
print('Total',total_survived_in_Percent.round(), '% were Survived in Titanic.')
```

### **Which age group, survived the most, kids (0-12), teen agers (13-19), youngsters (20-30), middle age people (31-50) or elders (50 and) greater) ?**

```python
Total_survived_age = Total_survived_age.count()
Total_survived_age
```

#### Total Survived **Kids**

```python
Total_survived_kidz = Total_survived_age[Total_survived_age["Age"].between(0, 12)]
Total_survived_kidz = Total_survived_kidz['Age'].count()
print('Total ',Total_survived_kidz, 'Kids with age group Of ( 0 to 12 ) was survived.')
```

#### Total Survived **Teen Agers**

```python
Total_survived_teen = Total_survived_age[Total_survived_age["Age"].between(13, 19)]
Total_survived_teen = Total_survived_teen['Age'].count()
print('Total ',Total_survived_teen, 'teen agers with age group Of ( 13 to 19 ) was survived.')
```

#### Total Survived **Youngsters**

```python
Total_survived_youngs = Total_survived_age[Total_survived_age["Age"].between(20, 30)]
Total_survived_youngs = Total_survived_youngs['Age'].count()
print('Total ',Total_survived_youngs, 'youngsters with age group Of ( 20 to 30 ) was survived.')
```

#### Total Survived **Middle Age**

```python
Total_survived_middle = Total_survived_age[Total_survived_age["Age"].between(31, 50)]
Total_survived_middle = Total_survived_middle['Age'].count()
print('Total ',Total_survived_middle, 'middle age with age group Of ( 31 to 50 ) was survived.')
```

#### Total Survived **Elders**

```python
Total_survived_elder = Total_survived_age[Total_survived_age["Age"] > 50]
Total_survived_elder = Total_survived_elder['Age'].count()
print('Total ',Total_survived_elder, 'elders with age group Of ( 50 or greater ) was survived.')
```
